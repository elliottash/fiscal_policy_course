```python
import pandas as pd 
import numpy as np 
df = pd.read_csv('state-tax-govt-data.csv')
df['log_gdp'] = np.log(df['gsp_q'])
df['log_incometax'] = np.log(df['individual_income_tax'])
keep = pd.notnull(df['log_gdp'] ) & np.isfinite(df['log_incometax'])
df = df[keep]
```

```python
df.head()
```

```python
df['_const'] = 1
```

```python
from linearmodels.iv import IV2SLS
```

```python
# OLS
formula_ols = 'log_gdp ~ 1 + C(state) + C(year) + log_incometax'
ols = IV2SLS.from_formula(formula_ols, data=df)
results_ols = ols.fit(cov_type='clustered',clusters=df['state'])
results_ols
```

```python
# First Stage
formula_fs = 'log_incometax ~ 1 + C(state) + C(year) + vote_nec_any_tax_inc'
fs = IV2SLS.from_formula(formula_fs, data=df)
results_fs = fs.fit(cov_type='clustered',clusters=df['state'])
results_fs
```

```python
# Reduced Form
formula_rf = 'log_gdp ~ 1 + C(state) + C(year) + vote_nec_any_tax_inc'
rf = IV2SLS.from_formula(formula_rf, data=df)
results_rf = rf.fit(cov_type='clustered',clusters=df['state'])
results_rf
```

```python
# 2SLS
formula_iv = 'log_gdp ~ 1 + C(state) + C(year) + [log_incometax ~ vote_nec_any_tax_inc]'
iv = IV2SLS.from_formula(formula_iv, data=df)
results = iv.fit(cov_type='clustered',clusters=df['state'])
results
```
