```python
# open data set and drop missing data.
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
df = pd.read_csv('state-tax-govt-data.csv')
df['log_deathtax'] = np.log(1+df['death_and_gift_tax'])
```

```python
# Governor Political Party
# Zero = Republican, One = Democrat
df.govparty_a.value_counts()
```

```python
df.lastdemmargin.hist(bins=100)
plt.show()
```

```python
keep = pd.notnull(df['log_deathtax']) & ((df.govparty_a == 0) | (df.govparty_a == 1)) & pd.notnull(df['lastdemmargin'])
df = df[keep]
```

```python
keep10 = np.abs(df['lastdemmargin']) < 10
df[keep10].lastdemmargin.hist(bins=100)
plt.show()
```

```python
# First Stage Viz
df['lastdemmargin_5pct'] = df['lastdemmargin'] - df['lastdemmargin'] % 5
df.groupby('lastdemmargin_5pct')[['govparty_a']].mean().plot()
```

```python
# Reduced Form Viz
df.groupby('lastdemmargin_5pct')[['log_deathtax']].mean().plot()
```

```python
# Reduced Form Viz
df['log_incometax'] = np.log(1+df['individual_income_tax'])
df.groupby('lastdemmargin_5pct')[['log_incometax']].mean().plot()
```

```python
# Run standard OLS model
import statsmodels.formula.api as smf # for running regressions
model = smf.ols('log_deathtax ~ govparty_a', data=df)
results = model.fit(cov_type='cluster',cov_kwds = {'groups': df['state']})
results.summary()
```

```python
# add polynomials in vote share
df['marginsq'] = df['lastdemmargin'] ** 2
model = smf.ols('log_deathtax ~ govparty_a + lastdemmargin + marginsq', data=df)
results = model.fit(cov_type='cluster',cov_kwds = {'groups': df['state']})
results.summary()
```

```python
# local regression around cutoff
keep5 = np.abs(df['lastdemmargin']) < 5
model = smf.ols('log_deathtax ~ govparty_a ', data=df[keep5])
results = model.fit(cov_type='cluster',cov_kwds = {'groups': df[keep5]['state']})
results.summary()
```
